```python
temp = np.array([36.5,37.2,0,38.1,0,36.8])
#zeros represent missing values
condition = temp != 0
filtered = temp[condition] #gathering of != temps in filtered
average_temp = filtered.mean()

clean_temp = np.where(temp == 0, average_temp,temp)
print(clean_temp)
```

```python
#print temps > 37
condition2 = clean_temp > 37
filtered2 = clean_temp[condition2]
print(filtered2)
```
