```python
import numpy as np
```

```python
patients_data = np.array([
[45, 80, 200, 120],
[60, 90, 240, 140],
[30, 70, 180, 110],
[50, 85, 220, 130],
[65, 95, 260, 150]
])

blood_pressure_values = patients_data[ : , 3: ]
cholesterol_level = patients_data[ : , 2:3]
high_risk = np.where((blood_pressure_values>140)| (cholesterol_level>240),"high_risk","normal")
print(high_risk)
```

```python
all_ages = patients_data[ : , 0:1]
olde_patient = all_ages >= 50
average_heart_rate_upove50 = olde_patient.mean()
average_heart_rate_upove50
```

```python
under50_patient = all_ages < 50
average_heart_rate_under50 = under50_patient.mean()
average_heart_rate_under50
```

```python
## What do you notice? 
## average_heart_rate_under50 < average_heart_rate_upove50
```
